# Getting non frequent words

```python
frequent_words = set(line.split('\t')[1] for line in lines[1:])
```

```python
def clean_text(text):
    tokens = word_tokenize(text)
    words = [word.lower() for word in tokens if word.isalpha() ]
    return words
```

```python
unknown_words = defaultdict(list)
n = last_page - first_page + 1
for i in tqdm(range(first_page, last_page + 1), total=n):
    page = doc.loadPage(i)
    text = page.getText("text")
    text_list = clean_text(text)
    unknown_words[i] = list( word for word in text_list if not word in frequent_words)
```

```python
print(len(unknown_words))
for key in unknown_words.keys():
    print(key, '-' * 80)
    print(' '.join(unknown_words[key]))
```
