```python
from ebooklib import epub
from PyPDF2 import PdfFileReader
import fitz


import os
```

```python
fpath = os.path.join('pdf', 'crimenYCastigo.pdf')
```

```python
doc = fitz.open(fpath)
print(f'Pages: {doc.pageCount}')
print(doc.metadata)
```

```python
first_page = 7
last_page = 563
i = 7
page1 = doc.loadPage(i)
page1text = page1.getText("text")
print(page1text)
```

```python
book = epub.read_epub(fpath)
```

```python
for i, doc in enumerate(book.get_items()):
    content = str(doc.content)
    print(i, '-' * 50)
    print(content)
    print(doc.)
```

```python
with open(fpath, 'rb') as fp:
    pdf = PdfFileReader(fp)
    info = pdf.getDocumentInfo()
    pages = pdf.getNumPages()
    
    i = 1
    page = pdf.getPage(i)
#     print(page)
    print(page.extractText())
```
